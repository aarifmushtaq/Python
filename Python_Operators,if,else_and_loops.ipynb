```python
#Arthematiuc Operators

a,b=7,6

print(a+b)
print(a-b)
print(a%b)
print(a/b)
print(a//b)
print(a**b)
```

```python
#Relational Operators

print(a>b)
print(a<b)
print(a==b)
print(a<=b)
print(a>=b)
print(a!=b)
```

```python
#Logical Operators
print(1 and 0)
print(1 or 0)
print(not 1)
```

```python
#Bitwise Operators

print (2 & 3)
print (2 | 3)
print (2 ^ 3)
print (~ 3)
print (4 >> 2)
print (4 << 1)

```

```python
#Asignment Operators

x=2
y=3
print(x,y)
x+=2
y+=3
print(x,y)
```

```python
#Membership Operators
#in and not in

print('D' in "Dehli")

print('d' in "Dehli")
print('d'  not in "Dehli")

x=[3,4,5,6,6,7,3,3]
print (3 in x)
print (3 not in x)
print (1 in x)
print (1 not in x)
```

```python
#Program To find su, of digits of a three digit number withoiut using loops

x=int(input("Enter The Number: "))

sum=0

lastdigit=x%10
x//=10
sum+=lastdigit

lastdigit=x%10
x//=10
sum+=lastdigit

lastdigit=x%10
x//=10
sum+=lastdigit

print(f"Sum Of Digits will be: {sum}")
```

```python
#if else

age=int(input("Enter The Age: "))
if age>=18:
  print("Eligible For Vote")
else:
  print("Not Eligible")

```

```python
#branched if else

#largest Of Three Numbers

x,y,z=6,10,1

if x>y and x>z:
  print(f"x is larger: {x}")
elif y>z:
  print(f"y is larger: {y}")
else:
  print(f"z is larger: {z}")
```
