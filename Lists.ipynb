```python
#nested if with list comprehension
basket=['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']

#add new list from myfruits and items if fruit in baskit and also starts with a
# nlist=[i for i in basket if i in my_fruits and i.startswith('a')]
nlist=[i for i in basket if i in my_fruits if i.startswith('a')]
print(nlist)
```

```python
#print 3 * 3 matrix
l=[[i for i in range (1,4) ] for i in range (1,4)]
print(l)
```

```python
#cartesian product
l1=[1,2,3,4]
l2=[5,6,7,8]
l3=[i*j for i in l1 for j in l2]
print(l3)
```
