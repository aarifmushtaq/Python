```python
#lists
a=5
b=9
print("Adress Of a & b")
print(id(a), id(b))

li=['apple',4,True,print]
print(li)

#How List Is Stored In Memory
l=[4,6,8,4,3,2]
print("Adress Of List")
print(id(l))


print("Adress Of list members")
for i in l:
  print(id(i))
```

```python
#Creating A List
a=[]  #empty list
b=[1,2,4,5,6,]  #homogeneous list
c=['a','apple',1,4] #heterogenous list
d=[1,2,[2,4,6]]  # 2d and Heterogenous
e=[[1,2,[3,4]],[3,5,4,[3,'li']]]  # 3d and homogenous
f=list('apple') #type conversion

print(a,b,c,d,e,f)
```

```python
l=[1,4,5,6,3,4,4,55]
for i in l:
  print(i,end=" ")

print("\nUsing Positive Indexing")
print("Index  Element")
for i in range(len(l)):
  print(i,"     ",l[i])

print("\nUsing Negative Indexing")
print("Index  Element")
for i in range(-1,-len(l),-1):
  print(i,"     ",l[i])

```

```python
#Editing

l=[1,2,4,'app','a',4,7,8]

#by indexing
print(l)
l[2]='Orange'
print(l)

#by Silicing
l[4:]=[1,2,3,4,5]
print(l)


```

```python
#deleting

l=[1,2,4,'app','a',4,7,8]
print(l)

del l[0]
print(l)

del l[4:]
print(l)

# del l
# print(l)

l.remove('app')
print(l)

l.pop()
print(l)

l.pop(0)
print(l)

l.clear()
print(l)

del l
```

```python
#Operations

# +

l1=[1,2,3,4]
l2=[3,4,5,6,7,7]
l3=l1+l2

print((f"{l1} and {l2} \n{l3}"))

# *

l4=l1*3
print(l4)

#memebership
print(1 in l1)
print('apple' not in l1 )

#loop
for i in l1:
  print(i)
```

```python
#list Functions

l1=['apple','orange','mango','pear']
l2=[1,4,6,3,2,4,5,8,9,2]

print(l1)
print(l2)

print(len(l1),len(l2))

print(sorted(l1,reverse=True) , sorted(l2,reverse=True))


print(max(l1),max(l2))

print(min(l1),min(l2))

print("-------------------------")

for i in l2:
  print(f"{i} : {l2.count(i)} ")

l1.reverse() #Reverse permentally
print(l1)

l2.sort() #sort permentally
print(l2)

l3=l1.copy() #shallow copy
print(l1)
print(id(l1),id(l3))

print(l2.index(2)) # retuens first index in case of multiplke same items

```

```python
#list Comprehension
l1=[1,2,3,4]
l2=[5,6,7,8]
l3=[i+j for i in l1 for j in l2] # second for will treated as nested
print(l1,l2,l3)
```

```python
#nested if with list comprehension
basket=['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']

#add new list from myfruits and items if fruit in baskit and also starts with a
# nlist=[i for i in basket if i in my_fruits and i.startswith('a')]
nlist=[i for i in basket if i in my_fruits if i.startswith('a')]
print(nlist)
```

```python
#print 3 * 3 matrix
l=[[i for i in range (1,4) ] for i in range (1,4)]
print(l)
```

```python
#cartesian product
l1=[1,2,3,4]
l2=[5,6,7,8]
l3=[i*j for i in l1 for j in l2]
print(l3)
```

```python
#zip Function
l1=[1,2,3,4]
l2=[5,6,7,8,9]

a=zip(l1,l2)
print(zip(l1,l2))

l3=[i+j for i,j in a]
print(l3)
```
