```python
#Count The frequnecy of character in a string without using the count func

s=input("Enter The String: ")
c=input("Enter The Character: ")
count = 0;

for i in s:
  if i==c:
    count+=1

print("Frequency: ",count)
```

```python
#convert the string into title case with using title()
s=input("Enter The string: ")

l=[]
d=s.split()
for i in d:
  l.append(i[0].upper() + i[1:].lower())
x=" ".join(l)
del l,d
print("After The Completion: ",x)
```
