```python
#ASSESSING THE STRINGS

#Positive Indexing
s='aarif mushtaq'
print(s[0], s[1],s[4])


#Negative indexing
print(s[-1],s[-5])



#Slicing

print(s[1:13])
print(s[:13])
print(s[1:])
print(s[1:5:1])
print(s[8:5:-1])
print(s[::-1])#Reverse
print(s[-7:-2:1])
print(s[2:8:-2])#first should be greater than seconmd while stepping is negative
```

```python
#Operations on Strings

a='aarif'
b='mushtaq'

#Arthematic operations
print(a+b)
print(a*8)

#Relational  Operations
#based On Lexiography
print(a==b)
print(a>b)
print(a!=b)
print(a<b)

#Logical Operators
print(a or b)
print(a and b)
print("" and b)
print("" or b)
print(not a)


#loops On strings

for i in a:
  print(a + " " + b)


#Memebershhiup Operator

print('a' in a )
print('A'  in a)
```
