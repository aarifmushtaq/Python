### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
x=int(input("Enter The Number: "))
sum=0
i=1
while i<=x:
  if i%5==0:
    i+=1
    continue
  sum=sum+i

  if sum>300:
    sum-=i
    break
  i+=1
print(sum)
````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
x=0
y=0

m="!"
print(
    "Enter 1 for UP: ",
    "\nEnter 2 for DOWN: ",
    "\nEnter 3 for LEFT: ",
    "\nEnter 4 for RIGHT: ",
    "\n Enter ! for stop"
)
m=input("Enter The Direction: ")
while m!="!":

  if m=="UP":
    x+=int(input("Enter The Steps: "))
  elif m=="DOWN":
    x-=int(input("Enter The Steps: "))
  elif m=="LEFT":
    y-=int(input("Enter The Steps: "))
  elif m=="RIGHT":
    y+=int(input("Enter The Steps: "))


  m=input("Enter The Direction: ")
  if m=="!":
    print("Robot Stopped")
    break
print(f"The Distance Is: {round((x**2+y**2)**0.5)}")

````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Write code here
x,y=int(input("Enter The First Number: ")),int(input("Enter The Second Number: "))
for i in range(x,y+1):
  sum=0
  for j in str(i):
    sum+=int(j)**3
  if sum==i:
    print(i)
````

###`Problem 14`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
# Write code here
#impttttttttttttttttttttttttttttt
import math
hour,minute=int(input("Enter The Hour: ")),int(input("Enter The Minute: "))
hour_angle = (hour * 30) + (minute * 0.5)
minute_angle = minute * 6
angle = abs(hour_angle - minute_angle)
angle = min(360 - angle, angle)
print(math.floor(angle))

````

A rectangle is denoted by providing the x and y coordinates of two points: the left top corner and the right bottom corner of the rectangle. Two rectangles sharing a side are considered overlapping. (L1 and R1 are the extreme points of the first rectangle and L2 and R2 are the extreme points of the second rectangle).

Note: It may be assumed that the rectangles are parallel to the coordinate axis.

````python
# Write code here
# (x1, y1) = L1  → left-top of rectangle 1
# (x2, y2) = R1  → right-bottom of rectangle 1

# (x3, y3) = L2  → left-top of rectangle 2
# (x4, y4) = R2  → right-bottom of rectangle 2
x1,x2=int(input("Enter The First X1: ")),int(input("Enter The Second X2: "))
y1,y2=int(input("Enter The First Y1: ")),int(input("Enter The Second Y2: "))
x3,x4=int(input("Enter The First X3: ")),int(input("Enter The Second X4: "))
y3,y4=int(input("Enter The First Y3: ")),int(input("Enter The Second Y4: "))

if x2 < x3 or x4 < x1 or y1 < y4 or y3 < y2:
    print("Rectangles do NOT overlap")
else:
    print("Rectangles overlap")
````
